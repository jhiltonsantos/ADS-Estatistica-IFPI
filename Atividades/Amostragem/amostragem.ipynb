# **Atividade de Amostragem**

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import itertools as it
from itertools import product
```

```
valor_populacao = np.array([1,2,4,5,7,9])

populacao = {
    'Media' :valor_populacao.mean(),
    'Variancia': valor_populacao.var(),
    'Tamanho Amostra': 3
}
```

## Visualizar valores da amostra

```
tabela_amostra = list(it.product(valor_populacao, repeat= populacao['Tamanho Amostra']))

data_tabela_amostra = pd.DataFrame(tabela_amostra)
data_tabela_amostra
```

## Visualizar valores das médias das amostras

```
media = []
for i in range(len(tabela_amostra)):
  media.append(sum(tabela_amostra[i]) / populacao['Tamanho Amostra'])
#media
data_media_amostra = pd.DataFrame(media)
data_media_amostra
```

## Visualizar valores de variancia correta das amostras

```
varianciac = []
for i in range(len(tabela_amostra)):
  varianciac.append(sum(tabela_amostra[i]) / populacao['Tamanho Amostra'])

data_variancia_certa_amostra = pd.DataFrame(varianciac)
data_variancia_certa_amostra
```

## Visualizar valores de variancia errada das amostras

```
# VarianciaE amostra
varianciae = []
for i in range(len(tabela_amostra)):
  varianciae.append(sum(tabela_amostra[i]) / populacao['Tamanho Amostra'])

data_variancia_errada_amostra = pd.DataFrame(varianciae)
data_variancia_errada_amostra
```

## Tabela População

```
print("Media: ", populacao['Media'])
print("Variancia: ", populacao['Variancia'])
print("Tamanho da Amostra: ", populacao['Tamanho Amostra'])
```

## Tabela Teória Estatística Média

```
teoria_media = {
    'Media': valor_populacao.mean(),
    'VarianciaTeoria' : populacao['Variancia'] / populacao['Tamanho Amostra'],
}

print("Media: ", teoria_media['Media'])
print("Variância teória: ", teoria_media['VarianciaTeoria'])
```

## Tabela Teória Estatística Variância

```
valor_amostral = []

for i in range(len(tabela_amostra)):
    valor_amostral += [np.var([tabela_amostra[i]])]

soma_valor_amostral = 0

for i in range(len(valor_amostral)):
    soma_valor_amostral += valor_amostral[i]

media_variancia_e = soma_valor_amostral/len(valor_amostral)

teoria_estatistica_variancia = {
    'Teoria Variancia' : populacao['Variancia'],
    'MediaVarianciaC': np.var(valor_amostral)/2,
    'MediaVarianciaE': media_variancia_e
}

print("Teoria Variância: ", teoria_estatistica_variancia['Teoria Variancia'])
print("Media VariânciaC: ", teoria_estatistica_variancia['MediaVarianciaC'])
print("Media VariânciaE: ", teoria_estatistica_variancia['MediaVarianciaE'])
```
